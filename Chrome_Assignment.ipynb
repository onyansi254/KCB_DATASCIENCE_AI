```python
# Analysing the salary by jobtitle and calculating the sum, min, max and mean of each.
grouped = df.groupby("experience_level").agg({"salary":["sum","min","max","mean"]})
grouped
```

```python
# Create subplots for each statistic
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Plot sum
axes[0, 0].bar(grouped.index, grouped["salary"]["sum"], color='blue')
axes[0, 0].set_title("Sum of Salaries")

# Plot min
axes[0, 1].bar(grouped.index, grouped["salary"]["min"], color='green')
axes[0, 1].set_title("Minimum Salary")
# Plot max
axes[1, 0].bar(grouped.index, grouped["salary"]["max"], color='red')
axes[1, 0].set_title("Maximum Salary")

# Plot mean
axes[1, 1].bar(grouped.index, grouped["salary"]["mean"], color='orange')
axes[1, 1].set_title("Average Salary")

# Set common labels and title
fig.suptitle("Salary Statistics by Experience Level")
fig.align_ylabels(axes)
plt.xlabel("Experience Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

For the sum graph, it shows that the seniors were earning the most while the entry_level people were earning the least.

The minimum salary for a senior level is 25000 while that of an entry-level is 14000.

For the maximum salaries, mid_level people seem to be earning the highest with the highest being 40000, while that of a senior is slightly lower and that of an entry level is 30000.

For the average salaries, the average salary for executive level is above 150000, while that of the senior level the average salary is 150000 while that of entrylevel is 90000.
